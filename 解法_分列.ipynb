## 2.0棋盘录入开局信息

```python
# 检查同一row的是否有相同的数

def reset_mutex_blog(index_num,num):
    #检查每一row/col/group
    row_num=int(sudoku_df.loc[index_num,'row_num'])
    col_num=int(sudoku_df.loc[index_num,'col_num'])
    group_key=str(sudoku_df.loc[index_num,'group_key'])
    index_list=sudoku_df[(sudoku_df['row_num']==row_num)|
                         (sudoku_df['col_num']==col_num)|
                         (sudoku_df['group_key'].str.contains(group_key))].index
    index_list=index_list.tolist()
    print('以下要更新index为：'+str(index_list))   
    
    #检查详细的每个区块
    num_chose='num_0'+str(num)
    sudoku_df.loc[index_list,num_chose]=0
    sudoku_df.loc[index_num,num_chose]=1


```

```python
# 将信息录入到棋盘df中
def input_start_blog(row,col,num):  
    
    index_num=sudoku_df[(sudoku_df['row_num']==row)&(sudoku_df['col_num']==col)].index    
    sudoku_df.loc[index_num,'num_real']=num
    
    #将信息录入df['num_01~09']中
  
    sudoku_df.loc[index_num,col_0_to_9]=0
    num_chose='num_0'+str(num)
    sudoku_df.loc[index_num,num_chose]=1   
    
    
    print(sudoku_df.loc[index_num])
```

## 4.0 检查并更新num_real

```python
#找到需要更新的blog
def reset_blog():
    print('now start to check')
    reset_df=sudoku_df[(sudoku_df[col_0_to_9].sum(axis=1)==1)&(sudoku_df['num_real']==0)]
    for index in reset_df.index.tolist():
        for num in col_0_to_9:
            if reset_df.loc[index,num]==1:
                num=int(num.replace('num_0',''))
                sudoku_df.loc[index,'num_real']=num
                #检查和此格互斥行和列
                reset_mutex_blog(index,num)
 
                
def re_check():
    reset_index=sudoku_df[(sudoku_df[col_0_to_9].sum(axis=1)==1)&(sudoku_df['num_real']==0)].index
    while len(reset_index)>0:
        print('now to re_check')
        reset_blog()
        reset_index=sudoku_df[(sudoku_df[col_0_to_9].sum(axis=1)==1)&(sudoku_df['num_real']==0)].index


```
