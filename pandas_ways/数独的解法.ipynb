## 2.0棋盘录入开局信息

```python
# 根据已经确定的blog更新相关row、col、group信息

def update_associated_blog(index_num,num):
    #检查每一row/col/group
    row_num=sudoku_df.loc[index_num,'row_num']
    col_num=sudoku_df.loc[index_num,'col_num']
    group_key=sudoku_df.loc[index_num,'group_key']
    index_list=sudoku_df[(sudoku_df['row_num']==row_num)|
                         (sudoku_df['col_num']==col_num)|
                         (sudoku_df['group_key']==group_key)].index
    index_list=index_list.tolist()
    print('后续将更新blog的index为：')
    print(str(index_list))
    
    #检查详细的每个区块
    for blog_index in index_list:

      
        blog_num_list=sudoku_df.loc[blog_index,'num_list']
       

     
        #判断blog——list是否为int
        if len(blog_num_list)==1:
            print('pass')
            
        else:
            if str(num) in blog_num_list:
                sudoku_df.loc[blog_index,'num_list']=blog_num_list.replace(str(num),'')
        print('-'*10)


```

```python
# 将信息录入到棋盘df中
def input_blog(row,col,num):
    index_num=sudoku_df[(sudoku_df['row_num']==row)&(sudoku_df['col_num']==col)].index    
    sudoku_df.loc[index_num,'num_real']=num
    sudoku_df.loc[index_num,'num_list']=str(num)    
    
```

## 3.0   更新棋盘信息

```python
#更新num——list
def update_table():
    check_indexs=sudoku_df[sudoku_df['num_real']!=0].index
    print('start to check if there has num_real to update')
    #检查每一个确定了数字的格子blog
    for index_num in check_indexs:
        
        num=sudoku_df.loc[index_num,'num_real']
        update_associated_blog(index_num,num)
        
```

```python
update_table()
```

```python
# 更新已经确定了的blog
def re_check():
    index_change=sudoku_df[(sudoku_df['num_list'].str.len()==1)&(sudoku_df['num_real']==0)].index
    while len(index_change)>0 :    
        for index_num in index_change:
            num=int(sudoku_df.loc[index_num,'num_list'])
            sudoku_df.loc[index_num,'num_real']=num
            update_associated_blog(index_num,num)
        index_change=sudoku_df[(sudoku_df['num_list'].str.len()==1)&(sudoku_df['num_real']==0)].index
```

```python
re_check()
```

```python
# 生成棋盘求解视图
def show_table(sudoku_df):
    mid_table=pd.pivot(sudoku_df,values='num_list',index=['row_num','group_key'],columns=['col_num'])
    return mid_table.fillna('')
    
```

## 4.0 找到互斥的twin——df

```python
# 将twins——df 相关blog 的num——list 进行更新
def pop_num_list_realateto_twins_df(col_name,num_list,unit_no):
    
    # 找到同一row/col/group 的 index，排除掉 twin——df 自己
    to_check_index=sudoku_df[(sudoku_df[col_name]==unit_no)&(sudoku_df['num_list']!=num_list)].index    
    sudoku_df.loc[to_check_index,'num_list']=(sudoku_df.loc[to_check_index,'num_list']).str.replace(num_list[0],'')
    sudoku_df.loc[to_check_index,'num_list']=(sudoku_df.loc[to_check_index,'num_list']).str.replace(num_list[1],'')
   
```

```python
def find_the_same_unit(the_df,num_list):
    
    
    
    #判断是否有相同的row——num
    if the_df[pd.DataFrame.duplicated(the_df,subset=['row_num'])].empty:
        print('no find same row_num')
    else:
        #找到row_num 相同的
        the_df=the_df[pd.DataFrame.duplicated(the_df,subset=['row_num'],keep=False)]
        unit_no=the_df['row_num'].iloc[0]
        col_name='row_num'
        pop_num_list_realateto_twins_df(col_name,num_list,unit_no)
        
     #判断是否有相同的col——num
    if the_df[pd.DataFrame.duplicated(the_df,subset=['col_num'])].empty:
        print('no find same col_num')
    else:
        #找到row_num 相同的
        the_df=the_df[pd.DataFrame.duplicated(the_df,subset=['col_num'],keep=False)]
        unit_no=the_df['col_num'].iloc[0]
        
        col_name='col_num'
        pop_num_list_realateto_twins_df(col_name,num_list,unit_no)
    
    #判断是否有相同的group_key
    if the_df[pd.DataFrame.duplicated(the_df,subset=['group_key'])].empty:
        print('no find same group')
    else:
        #找到group——key 相同的
        the_df=the_df[pd.DataFrame.duplicated(the_df,subset=['group_key'],keep=False)]
        unit_no=the_df['group_key'].iloc[0]
        col_name='group_key'
        pop_num_list_realateto_twins_df(col_name,num_list,unit_no)
```

```python
def recheck_twins_df():
    # 找到num——list 只有两种可能性的df
    two_num_df=sudoku_df[sudoku_df['num_list'].str.len()==2]
    #找到成对出现的twins_df(也可能有3个)
    count_df=two_num_df.groupby('num_list').count()
    twin_df_numlist=count_df[count_df['row_num']>1].index

    # 查询成对出现的df 是否有相同的 row/col/group
    for num_list in twin_df_numlist:
        the_df=two_num_df[two_num_df['num_list']==num_list]
        print(the_df)
        find_the_same_unit(the_df,num_list)
```
